```python
import pandas as pd
import numpy as np

df = pd.read_csv("base_sub_score_df.csv")
df.head()
```

```python
base_qol_cols = [col for col in df.columns if col.startswith("base_qol")]
sub_qol_cols = [col for col in df.columns if col.startswith("sub_qol")]
base_qol_df = df[base_qol_cols].rename(columns=lambda x: x.replace("base_qol_", ""))
sub_qol_df = df[sub_qol_cols].rename(columns=lambda x: x.replace("sub_qol_", ""))
```

```python
base_nan_indices = base_qol_df.isna()
sub_nan_indices = sub_qol_df.isna()
nan_indices = base_nan_indices | sub_nan_indices
```

```python
fu_df = pd.DataFrame()
fu_df["PCS"] = [b-a for a, b in zip(base_qol_df["PCS"], sub_qol_df["PCS"])]
fu_df["MCS"] = [b-a for a, b in zip(base_qol_df["MCS"], sub_qol_df["MCS"])]
fu_df["PKD"] = [b-a for a, b in zip(base_qol_df["PKD"], sub_qol_df["PKD"])]
fu_df["EKD"] = [b-a for a, b in zip(base_qol_df["EKD"], sub_qol_df["EKD"])]
fu_df["BKD"] = [b-a for a, b in zip(base_qol_df["BKD"], sub_qol_df["BKD"])]
```

```python
fu_df
```

```python
fu_df = fu_df > 0
fu_df = fu_df.where(~nan_indices, np.nan)
fu_df
```
